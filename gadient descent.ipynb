```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function y = (x + 3)^2
def function(x):
    return (x + 3) ** 2

# Define the derivative of the function, which is needed for gradient descent
def derivative(x):
    return 2 * (x + 3)
```

```python
# Initialize starting point (initial guess)
x = 2

# Set learning rate (controls the step size in each iteration)
learning_rate = 0.1

# Set the number of iterations
iterations = 20

# List to store x values and function values for visualization
x_values = []
y_values = []
```

```python
# Gradient Descent Loop
for i in range(iterations):
    # Append current x and function(x) to lists for visualization
    x_values.append(x)
    y_values.append(function(x))
    
    # Update x by moving against the gradient direction
    x = x - learning_rate * derivative(x)

# Print the final x value, which should be close to the local minimum
print("Local minimum occurs at x =", x)
print("Minimum value of function at this point y =", function(x))
```

```python
# Generate a range of x values for plotting the function curve
x_range = np.linspace(-5, 5, 100)
y_range = function(x_range)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="y = (x + 3)^2", color="blue")

# Plot the gradient descent path
plt.scatter(x_values, y_values, color="red", label="Gradient Descent Path")
plt.plot(x_values, y_values, color="red", linestyle="--", alpha=0.5)

# Annotate the initial and final points
plt.scatter(x_values[0], y_values[0], color="green", marker="o", label="Start Point")
plt.scatter(x_values[-1], y_values[-1], color="purple", marker="x", label="Local Minimum")

# Labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minimum of y = (x + 3)^2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def function(x):
    return (x + 5) ** 2

def gradient_descent(function, start, learn_rate, n_iter):
    x = start
    for _ in range(n_iter):
        grad = 2 * (x + 5)  # Derivative of the function (x+5)**2 with respect to x
        x = x - learn_rate * grad
        print(f"Iteration {_ + 1}: x = {x}, function(x) = {function(x)}")

gradient_descent(function=function, start=3.0, learn_rate=0.2, n_iter=50)
```
